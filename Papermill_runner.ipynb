```python
import papermill as pm
import subprocess
import ipywidgets as widgets
from IPython.display import display
import os
import shutil
```

```python
p = subprocess.Popen("pwd", stdout=subprocess.PIPE, shell=True)
pwd = p.stdout.read().decode("ascii").strip()
taxon = "\'Escherichia coli\'"
taxon_file = taxon.lower().replace(" ", "_").replace("\'", "")
number_of_genomes=10
```

```python
os.chdir(pwd)
if os.path.exists(taxon_file):
    shutil.rmtree(taxon_file)
os.mkdir(path=taxon_file,mode=0o777)
pwd2 = pwd + f"/{taxon_file}"
os.chdir(pwd2)
os.mkdir(path="gff_files",mode=0o777)
os.mkdir(path="annotated_genomes",mode=0o777)
```

```python
uploader = widgets.FileUpload(
    accept='',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=True  # True to accept multiple files upload else False
)
display(uploader)
```

```python
os.chdir(pwd2 + "/annotated_genomes")
for genome in uploader.value:
    with open(f"./{genome}", "wb") as fp:
        fp.write(uploader.value[genome]['content'])
os.chdir(pwd)
```

```python
pm.execute_notebook(
   pwd+'/MastersProject.ipynb',
   pwd2+f'/{taxon_file}_project.ipynb',
   #kernel_name='roary',
   parameters=dict(taxon=taxon, pwd=pwd2, number_of_genomes=number_of_genomes)
)
#Rhodothermus and Thermoanaerobacterium, "\'Deinococcus radiodurans\'"
```
